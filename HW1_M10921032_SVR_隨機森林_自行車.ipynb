```
#匯入模組
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.feature_extraction import DictVectorizer
```

```
#讀取bike資料
bike_data = pd.read_csv('drive/My Drive/Yuntech109_1/DataMining/Homework02/hour.csv')
```

```
bike_data
```

```
bike_data_df = pd.DataFrame(bike_data)
```

```
bike_data_df
```

```
bike_data_df.columns
```

```
# 把時間欄位轉為時間格式，在後面就可以直接取「月份、日期、小時」
bike_data_df['dteday'] = pd.to_datetime(bike_data_df['dteday'])
```

```
bike_data_df['year'] = bike_data_df['dteday'].dt.year
bike_data_df['month'] = bike_data_df['dteday'].dt.month
bike_data_df['date'] = bike_data_df['dteday'].dt.day
```

```
bike_data_df
```

```
bike_data_df.columns
```

```
bike_data_df = bike_data_df[['instant', 'year', 'month', 'date', 'dteday', 'season', 'yr', 'mnth', 'hr', 'holiday', 'weekday',
       'workingday', 'weathersit', 'temp', 'atemp', 'hum', 'windspeed',
       'casual', 'registered', 'cnt']]
```

```
bike_data_df
```

```
bike_data_df = bike_data_df.drop(['dteday'], axis=1)
```

```
bike_data_df
```

```
bike_data_df.to_csv('drive/My Drive/Yuntech109_1/DataMining/Homework02/bikehour.csv', index=False)
```

```
new_bike_data = pd.read_csv('drive/My Drive/Yuntech109_1/DataMining/Homework02/bikehour.csv')
```

```
new_bike_data
```

```
new_bike_data.columns
```

```
X = new_bike_data.iloc[:,0:18]
```

```
X
```

```
Y = new_bike_data.iloc[:,18]
```

```
Y
```

```
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2)
```

```
print("Xtrain")
print(X_train)
print("Xtest")
print(X_test)
print("Ytrain")
print(y_train)
print("Ytest")
print(y_test)
```

```
regr = make_pipeline(StandardScaler(), SVR(C=1.0, epsilon=0.2))
#regr = SVR()
```

```
regr_bike = regr.fit(X_train, y_train)
```

```
y_predict_svr = regr_bike.predict(X_test)
```

```
regr_bike.score(X_train,y_train)
```

```
regr_bike.score(X_test,y_test)
```

```
from math import sqrt
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

errors = abs(y_predict-y_test)
mape = 100 * (errors / y_test)
accuracy = np.mean(mape)

print('MAPE:', round(accuracy, 2), '%.')

print("RMSE:", round(sqrt(mean_squared_error(y_test,y_predict)),2), '%')
```

```
from math import sqrt
errors = abs(y_predict-y_test)
mape = 100 * (errors / y_test)
accuracy = np.mean(mape)

rms = sqrt(mean_squared_error(y_test, y_predict))

print('MAPE:', accuracy)
print('RMSE:', rms)
```
